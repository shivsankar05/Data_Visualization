 Using the given x and y data, generate a 3D surface plot to visualize the function

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
import plotly.express as px
fig = px.scatter_3d(df, x='X', y='Y', z='Z')
fig.show()
```
